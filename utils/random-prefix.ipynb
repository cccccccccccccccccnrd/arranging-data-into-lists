```typescript
async function s(term: string) {
  const response = await fetch(`https://www.youtube.com/results?search_query=${term}`)
  const text = await response.text()
  return text
}

function extractIds(html: string): string[] {
  const regex = /\/watch\?v=([^"]+)"/g;
  const ids = html.match(regex)?.map(match => {
    const id = match.split('v=')[1]
    return id.split('\\')[0]
  })
  return [...new Set(ids)] || [];
}

function gen(length = 5) {
  const chars = '0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'

  return Array.from({ length }, () =>
    chars[Math.floor(Math.random() * chars.length)]
  ).join('')
}

function query(amount = 3) {
  return Array.from({ length: amount }, () => gen()).join('|')
}

async function init () {
  for (let i = 0; i < 5; i++) {
    const p = query(3)
    const text = await s(p)
    const ids = extractIds(text)
    console.log(p, ids, ids.length)
  }
}

init()
```
